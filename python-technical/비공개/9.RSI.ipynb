# 상대강도지수 (relative strength index, RSI)
주가의 상승과 하락의 강도를 표현하는 방법
$$ \frac{14일 상승폭}{14일 상승폭 + 14일 하락폭}
$$
- RSI가 50이하이면 하락세
- RSI가 50이상이면 상승세

20일 데이터로 RSI 구해보기

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from pykrx import stock

df = stock.get_market_ohlcv_by_date("20210311", "20210507", "005930")


```

```python
prev = df['종가'].shift(1)
cond_ris = df['종가'] > prev
cond_fal = df['종가'] < prev
df['ris'] = 0
df.loc[cond_ris, 'ris'] = df['종가'] - prev
df['fal'] = 0
df.loc[cond_fal, 'fal'] = prev - df['종가']
df
```

```python
au = df['ris'].rolling(14).sum()
ad = df['fal'].rolling(14).sum()

s = au / (au + ad) * 100
s
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from pykrx import stock

# df = stock.get_market_ohlcv_by_date("20210311", "20210507", "005930")
ohlc = go.Candlestick(     
    x = df.index,
    open = df.시가,
    high = df.고가,
    low = df.저가,
    close = df.종가,
    increasing_line_color = "rgba(255, 60, 150, 1)",
    decreasing_line_color = "rgba(60, 150, 255, 1)",
)

df_ma12 = df.종가.ewm(span=12, adjust=False).mean()
df_ma26 = df.종가.ewm(span=26, adjust=False).mean()

rsi = go.Scatter(
    x = df.index,
    y = s
)

fig = make_subplots(rows=2, cols=1, row_heights=[0.7, 0.3])
fig.add_trace(ohlc, row=1, col=1)
fig.add_trace(rsi, row=2, col=1)
fig.update_layout(
    xaxis_rangeslider_visible = False,
    plot_bgcolor = "rgba(250, 250, 250, 1)",
    height =600
)
fig.show()
```
